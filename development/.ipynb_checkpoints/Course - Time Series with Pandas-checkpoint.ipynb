```python
df = pd.read_csv("Course Data/starbucks.csv", index_col="Date", parse_dates=True)
df
```

```python
df.index
```

```python
# daily -> yearly
df.resample(rule="YE").mean()
```

```python
def first_day(entry):
    if len(entry): return entry.iloc[0];
```

```python
df.resample(rule="YE").apply(first_day)
```

```python
df["Close"].resample("ME").max().plot.bar()
```

```python
df.Close.plot(figsize=(12,5))
```

```python
df.Close.plot(figsize=(12,6))
df.rolling(window=60).mean()["Close"].plot()
```

```python
df["Close: 30 day mean"] = df.Close.rolling(window=30).mean()
df
```

```python
df.Close.expanding().mean().plot(figsize=(12,5))
```
